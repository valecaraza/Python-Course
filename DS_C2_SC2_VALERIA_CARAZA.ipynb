```python
# Importar las librerías pandas y numpy
import pandas as pd
import  numpy as np

# a: Extraer información del archivo
df_vacunas = pd.read_csv('/content/R2country_vaccinations.csv')
print('-' * 50)

# b: Mostrar la estructura y tipos de datos
print('Estructura inicial del DataFrame:')
print(df_vacunas.info())

# Convertir la columna 'date' al tipo de dato datetime64
df_vacunas['date'] = pd.to_datetime(df_vacunas['date'])

print('-' * 50)
print('\nEstructura del DataFrame después de la conversión:')
print(df_vacunas.info())

# c: Determinar la cantidad de vacunas aplicadas de cada compañía
vacunas_por_compania = df_vacunas.groupby('vaccines')['total_vaccinations'].sum()
print('-' * 50)
print('\nCantidad de vacunas aplicadas por compañía:')
print(vacunas_por_compania)
print('-' * 50)

# d: Obtener la cantidad de vacunas aplicadas en todo el mundo
total_vacunas_mundo = df_vacunas['total_vaccinations'].sum()
print('\nCantidad de vacunas aplicadas en todo el mundo:', total_vacunas_mundo)
print('-' * 50)

# e: Calcular el promedio de vacunas aplicadas por país
promedio_vacunas_pais = df_vacunas.groupby('country')['total_vaccinations'].mean()
print('\nPromedio de vacunas aplicadas por país:')
print(promedio_vacunas_pais)
print('-' * 50)

# f: Cantidad de vacunas aplicadas el día 29/01/21
vacunas_29_01_21 = df_vacunas[df_vacunas['date'] == '2021-01-29']['daily_vaccinations'].sum()
print('\nCantidad de vacunas aplicadas el día 29/01/21 en todo el mundo:', vacunas_29_01_21)
print('-' * 50)

# g: Crear dataframe 'conDiferencias'con datos originales y una columna derivada (diferencias) con las
# diferencias de aplicación entre las columnas 'daily_vaccinations' y 'daily_vaccinations_raw'
conDiferencias = df_vacunas.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
print("\nDataFrame 'conDiferencias' con la nueva columna 'diferencias':")
print(conDiferencias.head())
print('-' * 50)

# h: Obtener el periodo de tiempo entre el registro con fecha más reciente y el
# registro con fecha más antigua
fecha_mas_reciente = df_vacunas['date'].max()
fecha_mas_antigua = df_vacunas['date'].min()
periodo_tiempo = fecha_mas_reciente - fecha_mas_antigua
print("\nPeriodo de tiempo entre las fechas más recientes y más antiguas es:", periodo_tiempo)
print('-' * 50)

# i: Crear DataFrame 'conCantidad' con datos originales y una columna derivada
# 'canVan' con la cantidad de vacunas utilizasas cada día
conCantidad = df_vacunas.copy()
# La función lamba cuenta el número de vacunas separadas por la coma
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').str.len()
print("\nDataFrame 'conCantidad' con la nueva columna 'canVac':")
print(conCantidad.head())
print('-' * 50)

# j: Generar DataFrame 'antes20' con todos los registros antes del 20/12/20
antes20 = df_vacunas[df_vacunas['date'] < '2020-12-20']
print("\nDataFrame 'antes20' (registros antes del 20/12/20):")
print(antes20.head())
print('-' * 50)

# k: Generar DataFrame 'pfizer' con todos los registros donde se haya
# utilizado la vacuna Pfizer
pfizer = df_vacunas[df_vacunas['vaccines'].str.contains('Pfizer', na=False)]
print("DataFrame 'pfizer':")
print(pfizer.head())
print('-' * 50)

# l: Almacenar los dataframes conDiferencias, con Cantidad, antes20 y pfizer
# en un archivo de excel denominado resultadosReto.xlsx
with pd.ExcelWriter('resultadosReto3.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
print("\nLos Data Frames se han guardado en el archivo 'resultadosReto.xlsx' con éxito.")
```
